```python
import chess.pgn
import collections
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
import pycountry_convert
from geopy.geocoders import Nominatim
import folium
import cartopy.crs as ccrs
from tqdm import tqdm
```

```python
pgn = open("caissabase.pgn") # DB containing ~4 million chess games
```

```python
# VISUALIZATION 4---------------------------------------------------NUMBER OF MOVES PER GAME
moves = {}
pbar = tqdm(total=4270000)

while True:
    # read in game
    game = chess.pgn.read_game(pgn)
    
    # if reached end of pgn file
    if game is None:
        break
    
    board = game.board()
    game_length = 0
    
    # loop through all moves in this game
    for move in game.mainline_moves():
        game_length += 1
        board.push(move)
    
    try: # populate moves dictionary   
        moves[game_length] += 1
    except: # if no previous game with same number of moves, create new entry
        moves.update({game_length: 1})
    
    pbar.update(1) # update progress bar
    
pbar.close() # close progress bar
moves[0] = 0 # remove invalid games with 0 moves

# configure and plot bar chart
plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(20,14), dpi=800)
ax.grid(False)
ax.set_title('Frequencies of Game Length in Moves')
sns.set(font_scale = 2)
# plt.hist(moves, bins=len(moves), color="blue", edgecolor='blue')
plt.bar(list(moves.keys()), moves.values(), color='r')
plt.savefig('fig/moves_histogram.png')
plt.show()
```
